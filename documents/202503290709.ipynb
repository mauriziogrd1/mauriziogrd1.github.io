$\displaystyle\frac{1}{k}\sum_{i=n−k+1}^{n}x_i$

```python
window = 10

sma_values = []

for i in range(len(data['Close'])):
  if i < window - 1:
      sma_values.append(None)
  else:
      window_data = data['Close'][i - window + 1:i + 1]
      sma = sum(window_data) / window
      sma_values.append(sma)
```
